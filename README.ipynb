> **Git**

Git es un  sistema que hace control de versiones gratis de codigo abierto que se utiliza principalmente por desarrolladores para ver cambios en el codigo fuente propio con base a prototipos y versiones del codigo y puedan compartirlo entre otros desarrolladores y poder ayudarse entre todos.

> **Que contiene git**
Tiene herramientas para poder compartir el codigo principal para demas desarrolladores para poder tener de forma local en el sistema de controlador de versiones (VCS)
y se guarda en un historial de git llamado repositorio y contiene un COMMIT que es la forma en cual se guarda cada actualizacion del codigo y este puede contener RAMAS que hace que el codigo principal se pueda sacar una rama para desarrollar de forma separada y despues esa rama la puedes unir sin forma de dañar el proyecto unificnandola Y el CLON es la forma en cual cada desarrollador debe tener un clon esel repositorio local en cada pc.
-**VCS:** controlador de versones.
-**COMMIT** forma en la cual se guarda cada actualizacion del codigo en forma de versiones usadas por el desarrollador propio.
-**Ramas** Forma de que un desarrollador pueda agregar una funcion o demas cosas en el codigo de forma externa del original  unificarlo de forma segura sin miedo a dañar algun componente.
-**Clon:** Es la cual cada desarrollador debe un repositorio del codigo original guardado en cada pc local de cada uno.
-**Fork:** Proyecto diferente que se crea a travez de otro diferente y se basa en otro proyecto ya creado.

> **Creador de git**
Linus Torvalds, creador de Git, lo desarrolló en 2005 como respuesta a una crisis en el proyecto del kernel de Linux, buscando una herramienta de control de versiones rápida, distribuida y libre.

>**Principales comandos de git** \
GIT INIT: Inicializa un nuevo repositorio Git en el directorio actual.
GIT CLONE: Copia un repositorio remoto a tu máquina local.
GIT ADD: Añade archivos al área de preparación (staging area) para el próximo commit.
GIT COMIT: Guarda los cambios en el historial del repositorio con un mensaje.
GIT STATUS: Muestra el estado actual del repositorio (archivos modificados, etc.)
GIT BRANCH: Muestra, crea o elimina ramas.
GIT CHECKOUT: Cambia de rama o restaura archivos.
